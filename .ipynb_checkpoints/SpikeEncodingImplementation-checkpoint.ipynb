# Selection and Optimization of Temporal Spike Encoding Methods for Spiking Neural Networks

## $1^{st}$ temporal contrast encoding: <span style="color:red">Threshold-based representation (TBR)</span>

**Summary:** Compares the absolute signal change of an input signal with a threshold

**Advantages:**
1. Fast and simple
2. Reduce white noise by applying the threshold to small perturbations in the signal
3. No falsely registered values at the start or end of neither the spike train nor the reconstructed signal

**Disadvantages:**
1. Sensitive to strong white noise in the signal resulting in strong low-frequency artifact components
2. Tradeoff between representing small and large events. Since the whole sample length is considered for threshold value calculation, this encoding can be disadvantageous for long samples where there may be amplitude differences between events of different parts of the sample

<img src="./img/TBR_Encoding.png" width="400" height="400">

#### Load packages

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
from scipy import signal
```

#### Load encoding functions

```python
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w

def white_noise(rho, sr, n, mu=0):
    sigma = rho * np.sqrt(sr/2)
    noise = np.random.normal(mu, sigma, n)
    return noise

def sin_wave_generator(amp, f, n):
    time = np.linspace(0, p, n)
    signal = amp*np.sin(f*np.pi*time)
    return time,signal

def TBR_encoder(signal, factor):
    signalDiff    = []
    spike_output  = []
    for t in range(1,len(signal)-1):
        diff = signal[t+1] - signal[t]
        signalDiff.append(diff)
    signalDiff[-1] = signalDiff[len(signalDiff)-1]
    threshold = np.mean(signalDiff) + factor*np.std(signalDiff)
    for t in range(len(signal)):
        if signal[t] > threshold:
            spike = 1
            spike_output.append(spike)
        else:
            spike = -1
            spike_output.append(spike)
    return spike_output,threshold

def temporal_contrast(signal, factor):
    # Based on algorithm provided in:
    #   Sengupta et al. (2017)
    #   Petro et al. (2020)
    diff = np.zeros(len(data)-1)
    spikes = np.zeros(len(data))
    for i in range(len(data)-1):
        diff[i] = data[i+1] - data[i]
    threshold = np.mean(diff) + factor * np.std(diff)
    diff = np.insert(diff, 0, diff[1])
    for i in range(len(data)):
        if diff[i] > threshold:
            spikes[i] = 1
        elif diff[i] < -threshold:
            spikes[i] = -1
    return spikes, threshold

def spike_decoder(spike_output,threshold,signal):
    startpoint = signal[0]
    recon = np.zeros(len(spike_output))
    recon[0] = startpoint
    for t in range (2,len(spike_output)):
        if spike_output[t] == 1:
            recon[t] = recon[t-1] + threshold
        elif spike_output[t] == -1:
            recon[t] = recon[t-1] - threshold
        else:
            recon[t] = recon[t-1]
    return recon
```

#### Threshold-based representation encoding results on a sinusoidal periodic wave

```python
factor = 0.99 # TBR encoder factor
amp       = 1     # amplitude
rho       = 1     # spectral noise density unit/SQRT(Hz)
f         = 10    # frequency
n         = 500   # number of points
sr        = 100   # sample rate
p         = n/sr  # period

(time,signal) = sin_wave_generator(amp, f, n)
noise = white_noise(rho, sr, n)
# signal = signal + noise # to add noise to the raw signal

#### Method 1 ######
# (spike_output,threshold) = TBR_encoder(signal, factor)

#### Method 2 ######
(spike_output,threshold) = temporal_contrast(signal, factor)
recon = spike_decoder(spike_output,threshold,signal)

fig, axs = plt.subplots(3, 1,figsize=(15, 7))
axs[0].plot(time, signal)
axs[0].set_ylabel('voltage(V)')
axs[1].bar(time,spike_output,width=0.01,align='center')
axs[1].set_ylabel('Spikes')
axs[2].plot(time, signal)
axs[2].plot(time, recon,'r--')
axs[2].set_ylabel('voltage(V)')
axs[2].legend(['Original','Recons'])
plt.tight_layout()
plt.show()
```

## $2^{nd}$ temporal contrast encoding: <span style="color:red">Step-Forward (SF)</span>

**Summary:** Uses the next available signal value and checks if the previous value (base) and an additional threshold is exceeded

**Advantages:**
1. The threshold can be chosen such that small and large amplitude events are both well represented
2. Noise in the signal is minimally reduced by the threshold
3. Due to the moving baseline, the reconstructed signal does not drift away even for longer signals (No overshoot)

**Disadvantages:**
They havent reported any disadvantage elsewhere

#### Load encoding functions

```python
def SF_encoder(signal, threshold):
    start_point = signal[0]
    spike_output = np.zeros(len(signal))
    base = signal[0]
    for t in range(2,len(signal)):
        if signal[t] > base + threshold:
            spike_output[t] = 1
            base = base + threshold
        else:
            spike_output[t] = -1
            base = base - threshold
    return spike_output, start_point

def step_forward(signal, threshold):
    # Based on algorithm provided in:
    #   Petro et al. (2020)
    startpoint = data[0]
    spikes = np.zeros(len(data))
    base = startpoint
    for i in range(1,len(data)):
        if data[i] > base + threshold:
            spikes[i] = 1
            base = base + threshold
        elif data[i] < base - threshold:
            spikes[i] = -1
            base = base - threshold
    return spikes, startpoint
```

## $3^{rd}$ temporal contrast encoding: <span style="color:red">Moving window (MW)</span>

**Summary:** Uses a base which is defined by the mean of the previous signal in a time window

**Advantages:**
1. It is robust against white noise
2. Moving averge is an optimal smoothing filter in the time domain against white noise
3. Could be used to filter out a strong, artifact frequency component, e.g., power line noise in EEG signals

**Disadvantages:**
1. Window size heavily affect the cutoff frequency, the slope of the frequency response ad the noise reduction
2. The signal beginning is not encoded well until the window size is reached, therefore it requires tricks to adjust the baseline
3. Overshoot-type errors often appear

<img src="./img/MW_Encoding.png" width="400" height="400">

#### Load encoding functions

```python
def MW_encoder(signal, threshold, window):
    start_point = signal[0]
    spike_output = np.zeros(len(signal))
    base = np.mean(signal[0:window+1])
    for t in range(1,window+1):
        if signal[t]>base+threshold:
            spike_output[t] = 1
        else:
            spike_output[t] = -1
    for t in range(window+2,len(signal)):
        base = np.mean(signal[t-window-1:t-1])
        if signal[t]>base+threshold:
            spike_output[t] = 1
        else:
            spike_output[t] = -1
    return spike_output

def moving_window(data, threshold, window):
    # Based on algorithm provided in:
    #   Petro et al. (2020)
    startpoint = data[0]
    spikes = np.zeros(len(data))
    base = np.mean(data[0:window+1])
    for i in range(window+1):
        if data[i] > base + threshold:
            spikes[i] = 1
        elif data[i] < base - threshold:
            spikes[i] = -1
    for i in range(window+2, len(data)):
        base = np.mean(data[(i-window-1):(i-1)])
        if data[i] > base + threshold:
            spikes[i] = 1
        elif data[i] < base - threshold:
            spikes[i] = -1
    return spikes, startpoint
```

#### Moving average encoding results on a sinusoidal periodic wave

```python
window    = 5     # MW encoding parameters
threshold = 0.2   # MW encoding parameters
amp       = 1     # amplitude
rho       = 1     # spectral noise density unit/SQRT(Hz)
f         = 10    # frequency
n         = 500   # number of points
sr        = 100   # sample rate
p         = n/sr  # period

(time,signal) = sin_wave_generator(amp, f, n)

###### Method 1 ##########
# (spike_output) = MW_encoder(signal, threshold, window)

##### Method 2 ###########
(spike_output, startpoint) = moving_window(signal, threshold, window)

recon = spike_decoder(spike_output,threshold,signal)

fig, axs = plt.subplots(3, 1,figsize=(15, 7))
axs[0].plot(time, signal)
axs[0].set_ylabel('voltage(V)')
axs[1].bar(time,spike_output,width=0.01,align='center')
axs[1].set_ylabel('Spikes')
axs[2].plot(time, signal)
axs[2].plot(time, recon,'r--')
axs[2].set_ylabel('voltage(V)')
axs[2].legend(['Original','Recons'])
plt.tight_layout()
plt.show()
```

```python
def hough_spike(data, fir):
    # Based on algorithm provided in:
    #   Schrauwen et al. (2003)
    spikes = np.zeros(len(data))
    shift = min(data)
    data = data - shift*np.ones(len(data))
    for i in range(len(data)):
        count = 0
        for j in range(len(fir)):
            if i+j < len(data):
                if data[i+j] >= fir[j]:
                    count = count + 1
        if count == len(fir):
            spikes[i] = 1
            for j in range(len(fir)):
                if i+j < len(data):
                    data[i+j] = data[i+j] - fir[j]
    return spikes, shift

```
